# GTEx data analysis

```python
import pandas as pd
file_locate = "/data/" # Place the files to the folder "data"
```

```python
skin_annotation = pd.read_csv(filepath_or_buffer = file_locate+"GTEx_Analysis_v8_Annotations_SampleAttributesDS.txt", sep = "\t")
skin_columns = list(skin_annotation[skin_annotation['SMTSD'].str.contains("Skin")]['SAMPID'])
```

```python
path_to_file = file_locate+"GTEx_transcript_v8.gct"
with open(path_to_file) as input_file:
    head = [next(input_file) for _ in range(1)]
sample_transcript = head[0].split("\t")
sample_transcript[-1] = sample_transcript[-1].split("\n")[0]
```

```python
final_col = list(set(skin_columns).intersection(set(head[0].split("\t")[:-1]))) + sample_transcript[0:2]
skin_transcript = pd.read_csv(file_locate+"GTEx_transcript_v8.gct", sep = "\t", index_col=0, header=0, usecols = final_col)
```

```python
numb_transcript = skin_transcript.loc[skin_transcript["gene_id"].str.contains("ENSG00000133961")]
numb_transcript.to_csv(file_locate+"numb_skin_TPM.csv")
```

```python
plk5_transcript = skin_transcript.loc[skin_transcript["gene_id"].str.contains("ENSG00000185988")]
plk5_transcript.to_csv(file_locate+"PLK5_skin_TPM.csv")
```
